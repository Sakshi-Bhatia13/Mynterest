```python
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
```

```python
#sak code
from diffusers import StableDiffusionXLPipeline
import torch
from transformers.pipelines.image_to_text import Image
from IPython.display import display
```

```python
pipe = StableDiffusionXLPipeline.from_pretrained(
    "segmind/SSD-1B",
    torch_dtype = torch.float16,
    use_safetensors = True,
    cariant = 'fp16',
)
```

```python
pipe.to('cuda')
```

```python
import pandas as pd

# Load the CSV file
csv_path = '/content/outfit_descriptions_with_neg_prompts.csv'
prompts_df = pd.read_csv(csv_path)

# Assuming the CSV has a column named 'prompt' for positive prompts
# and a column named 'neg_prompt' for negative prompts
positive_prompts = prompts_df['Outfit_Description'].tolist()
negative_prompts = prompts_df['Negative_Prompt'].tolist()
```

```python
img_list = []
for i in range(100):
  image = pipe(prompt = positive_prompts[i], negative_prompt = negative_prompts[i]).images[0]
  img_list.append(image)
```

```python
import os
output_folder = '/content/generated_images'
os.makedirs(output_folder, exist_ok=True)

# Generate and save images
for i, (prompt, neg_prompt) in enumerate(zip(positive_prompts, negative_prompts)):
    try:
        image = pipe(prompt=prompt, negative_prompt=neg_prompt).images[0]
        image_path = os.path.join(output_folder, f'image_{i}.png')
        image.save(image_path)
        print(f'Saved image {i} to {image_path}')
    except Exception as e:
        print(f'Error generating image for prompt {i}: {e}')
```
